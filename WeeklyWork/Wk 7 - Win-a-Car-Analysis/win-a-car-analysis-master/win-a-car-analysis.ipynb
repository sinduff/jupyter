# Analysis of win a car

***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<br>

#### Read the data

***

```python
df = pd.read_csv("2020-11-01-win-a-car-data.csv")
df
```

<br>

#### How many people stayed with their original choice

***

```python
# We use a catplot to create a bar chart.
sns.catplot(x="stay", kind="count", data=df)
```

<br>

#### Random selection placing the car

***

```python
# Bar chart of where the car was placed.
sns.catplot(x="car", kind="count", data=df)
```

<br>

#### How many people selected different doors?

***

```python
# Bar chart of where the different picks.
sns.catplot(x="pick", kind="count", data=df)
```

<br>

#### Is changing a good idea?

***

```python
df["win"] = ((df["car"] == df["pick"]) & df["stay"]) | ((df["car"] != df["pick"]) & ~df["stay"])
```

```python
df
```

```python
# Bar chart of winning.
sns.catplot(x="win", kind="count", data=df)
```

```python
# Half the people happened to win.
df["win"].sum()
```

```python
# Bar chart of winning.
sns.catplot(x="stay", hue="win", kind="count", data=df)
```
